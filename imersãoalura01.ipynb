```python
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Snior',
    'MI': 'Pleno',
    'EN': 'Jnior',
    'EX': 'Executivo'
})

print("Categorias da coluna 'senioridade' traduzidas com sucesso!")
print(df['senioridade'].value_counts())
```

```python
df['porte_empresa'] = df['porte_empresa'].replace({
    'M': 'Mdio',
    'L': 'Grande',
    'S': 'Pequeno'
})

print("Categorias da coluna 'porte_empresa' traduzidas com sucesso!")
print(df['porte_empresa'].value_counts())
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Perodo',
    'FL': 'Freelancer'
})

print("Categorias da coluna 'contrato' traduzidas com sucesso!")
print(df['contrato'].value_counts())
```

```python
df['porte_empresa'] = df['porte_empresa'].replace({
    'M': 'Mdio',
    'L': 'Grande',
    'S': 'Pequeno'
})

print("Categorias da coluna 'porte_empresa' traduzidas com sucesso!")
print(df['porte_empresa'].value_counts())
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    50: 'Remoto',
    100: 'Hbrido'
})

print("Categorias da colunna 'remoto' traduzidas com sucesso!")
print(df['remoto'].value_counts())
```
