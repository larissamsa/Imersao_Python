```python
df.columns =[
    'ano',
    'senioridade',
    'contrato',
    'cargo',
    'salario',
    'moeda',
    'salario_usd',
    'residencia',
    'remoto',
    'empresa',
    'porte_empresa'
]

print(df.columns)
```

```python
df.head()
```

```python
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
})

print(df['senioridade'].value_counts())
```

```python
df['porte_empresa'] = df['porte_empresa'].replace({
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'

})

print(df['porte_empresa'].value_counts())
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
})

print(df['contrato'].value_counts())
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    100: 'Remoto',
    50: 'Híbrido'
})

print(df['remoto'].value_counts())
```

Criação de DataFrame abaixo


```python
#para colocar valores nulos
import numpy as np

#Criação de DataFrame de teste para exemplo
df_salarios = pd.DataFrame({
    'nome':  ['Ana','Bruno', 'Carlos', 'Daniele', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000 ]
})

'''criando nova coluna 'salarios_media' que será preenchida a apartir do cálculo da média com base nos salários e assim preenche o que tá nulo na nova coluna'''
df_salarios['salarios_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2)) #calcula a media e arredonda
df_salarios['salarios_media'] = df_salarios['salario'].fillna(df_salarios['salario'].median()) #calcula a mediana
df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

#ffill preenche com valor anterior (forward fill)
df_temperaturas['preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

#ffill preenche com o próximo valor (back)
df_temperaturas['Preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame ({
    'Nome':  ['Ana','Bruno', 'Carlos', 'Daniele', 'Val'],
    'Cidade': ['São Paulo', np.nan, 'Curitiba', np.nan, 'Belém']

})

#Prenchimento com texto 'Não informado' onde era nulo
df_cidades['Cidade_Preenchida'] = df_cidades['Cidade'].fillna('Não informado')
df_cidades
```
