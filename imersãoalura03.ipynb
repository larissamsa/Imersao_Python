```python
df.columns =[
    'ano',
    'senioridade',
    'contrato',
    'cargo',
    'salario',
    'moeda',
    'salario_usd',
    'residencia',
    'remoto',
    'empresa',
    'porte_empresa'
]

print(df.columns)
```

```python
df['senioridade'] = df['senioridade'].replace({
    'SE': 'SÃªnior',
    'MI': 'Pleno',
    'EN': 'JÃºnior',
    'EX': 'Executivo'
})

print(df['senioridade'].value_counts())
```

```python
df['porte_empresa'] = df['porte_empresa'].replace({
    'M': 'MÃ©dio',
    'L': 'Grande',
    'S': 'Pequeno'

})

print(df['porte_empresa'].value_counts())
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio PerÃ­odo',
    'FL': 'Freelancer'
})

print(df['contrato'].value_counts())
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    100: 'Remoto',
    50: 'HÃ­brido'
})

print(df['remoto'].value_counts())
```

CriaÃ§Ã£o de DataFrame abaixo


```python
#para colocar valores nulos
import numpy as np

#CriaÃ§Ã£o de DataFrame de teste para exemplo
df_salarios = pd.DataFrame({
    'nome':  ['Ana','Bruno', 'Carlos', 'Daniele', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000 ]
})

'''criando nova coluna 'salarios_media' que serÃ¡ preenchida a apartir do cÃ¡lculo da mÃ©dia com base nos salÃ¡rios e assim preenche o que tÃ¡ nulo na nova coluna'''
df_salarios['salarios_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2)) #calcula a media e arredonda
df_salarios['salarios_media'] = df_salarios['salario'].fillna(df_salarios['salario'].median()) #calcula a mediana
df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'TerÃ§a', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

#ffill preenche com valor anterior (forward fill)
df_temperaturas['preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'TerÃ§a', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

#ffill preenche com o prÃ³ximo valor (back)
df_temperaturas['Preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame ({
    'Nome':  ['Ana','Bruno', 'Carlos', 'Daniele', 'Val'],
    'Cidade': ['SÃ£o Paulo', np.nan, 'Curitiba', np.nan, 'BelÃ©m']

})

#Prenchimento com texto 'NÃ£o informado' onde era nulo
df_cidades['Cidade_Preenchida'] = df_cidades['Cidade'].fillna('NÃ£o informado')
df_cidades
```

```python
import plotly.express as px
```

```python
df_senioridade_salario = df_limpo.groupby('senioridade')['salario_usd'].mean().reset_index()

ordem_senioridade = ['JÃºnior', 'Pleno', 'SÃªnior', 'Executivo']

fig = px.bar(df_senioridade_salario,
             x='senioridade',
             y='salario_usd',
             title='MÃ©dia Salarial por Senioridade',
             labels={'senioridade': 'Senioridade', 'salario_usd': 'SalÃ¡rio MÃ©dio Anual (USD)'},
             category_orders={'senioridade': ordem_senioridade})

fig.update_layout(xaxis_title='Senioridade', yaxis_title='SalÃ¡rio MÃ©dio Anual (USD)')
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names= 'tipo_trabalho',
             values='quantidade',
             title= 'ProporÃ§Ã£o dos tipos de Trabalho'
             )
fig.show()
```
